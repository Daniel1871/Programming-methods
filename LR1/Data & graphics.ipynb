```python
import random
import numpy as np
import matplotlib.pyplot as plt
```

```python
f = open("surnames_table.jsonl", "r")
surnames = []
for line in f:
    surname = line.split('"')[3]
    try:
      if not(surname[-1] in 'ая'):
        surnames.append(surname)
    except:
        pass
```

```python
f = open("names_table.jsonl", "r")
names = []
for line in f:
    name = line.split('"')[3]
    try:
      if not(name[-1] in 'ая'):
        names.append(name)
    except:
        pass
```

```python
f = open("midnames_table.jsonl", "r")
midnames = []
for line in f:
    midname = line.split('"')[3]
    try:
      if not(midname[-1] in 'ая'):
        midnames.append(midname)
    except:
        pass
```

```python
len(surnames), len(names), len(midnames)
```

```python
sizes = [100, 1000, 5000, 10000, 20000, 50000, 100000]
```

```python
ranks = {'Рядовой' : range(18, 31),
        'Младший сержант' : range(18, 31),
        'Сержант' : range(18, 31),
        'Старший сержант' : range(22, 36),
        'Старшина' : range(30, 51),
        'Прапорщик' : range(30, 51),
        'Старший прапорщик' : range(30, 51),
        'Младший лейтенант' : range(30, 51),
        'Лейтенант' : range(30, 51),
        'Старший лейтенант' : range(30, 51),
        'Капитан' : range(35, 56),
        'Майор' : range(35, 56),
        'Подполковник' : range(40, 56),
        'Полковник' : range(40, 56)}
```

```python
for n in sizes:
    f = open(f"data{n}.txt", "w")
    for i in range(n):
        surname = random.choice(surnames)
        name = random.choice(names)
        midname = random.choice(midnames)
        rank = random.choice(list(ranks.keys()))
        number = random.randint(1, 8)
        age = random.choice(ranks[rank])
        if i < n - 1:
          f.write(f"{surname} {name} {midname};{rank};{number};{age}\n")
        else:
          f.write(f"{surname} {name} {midname};{rank};{number};{age}")
    f.close()
```

```python
select_time = [0.000128354, 0.0124257, 0.280602, 1.23768, 4.30609, 29.5927, 121.565]
insert_time = [8.3021e-05, 0.00727479, 0.180875, 0.743975, 3.04791, 19.7053, 89.563]
heap_time = [5.6286e-05, 0.00168614, 0.00517724, 0.0104426, 0.0294332, 0.065655, 0.167478]
```

```python
plt.plot(sizes, select_time, 'b', marker='o')
plt.plot(sizes, insert_time, 'r', marker='X')
plt.plot(sizes, heap_time, 'g', marker='*')
plt.grid(True)
plt.legend(['Select sort', 'Insert sort', 'Heap sort'])
plt.xlabel('Размер массива')
plt.ylabel('Время, секунды')
plt.title('Зависимости времени сортировки от размерности массива')
plt.show()
plt.show()
```

```python
plt.plot(sizes, np.log(select_time), 'b', marker='o')
plt.plot(sizes, np.log(insert_time), 'r', marker='X')
plt.plot(sizes, np.log(heap_time), 'g', marker='*')
plt.grid(True)
plt.legend(['Select sort', 'Insert sort', 'Heap sort'])
plt.xlabel('Размер массива')
plt.ylabel('Натуральный логарифм времени в секундах')
plt.title('Зависимости времени сортировки от размерности массива')
plt.show()
plt.show()
```
